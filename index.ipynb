```python
import pandas as pd
ruta="https://github.com/Enrique867/PC2/raw/refs/heads/main/HDI.xlsx"
data = pd.read_excel(ruta, engine='openpyxl', dtype=str,keep_default_na=False)
```

```python
print(data.shape)
data.head()
```

```python
columnas_a_eliminar = ['Unnamed: 3', 'Unnamed: 5', 'Unnamed: 7', 'Unnamed: 9', 'Unnamed: 11', 'Unnamed: 13']


data = data.drop(columns=columnas_a_eliminar, errors='ignore')


print(data.shape)
data.head()
```

```python
data.info()
```

```python
import re
start_row = 7
start_col = 3

def extraer_numero_con_signo(texto):
    if isinstance(texto, str):
        match = re.search(r'^[+-]?\d+(\.\d+)?', texto)
        return match.group(0) if match else None
    return None

for col in data.columns[start_col:]:
    data.loc[start_row:, col] = data.loc[start_row:, col].apply(extraer_numero_con_signo)


print(data.iloc[6:, 2:])
```

```python
start_row = 7
start_col = 3


for col_idx in range(start_col, len(data.columns)):
    col_name = data.columns[col_idx]
    data.loc[start_row:, col_name] = pd.to_numeric(data.loc[start_row:, col_name], errors='coerce')


print(data.shape)
data.head(15)
```

```python
nuevos_nombres = {
    data.columns[2]: 'HDI_Value',
    data.columns[3]: 'Life_Expectancy_at_Birth',
    data.columns[4]: 'Expected_Years_of_Schooling',
    data.columns[5]: 'Mean_Years_of_Schooling',
    data.columns[6]: 'GNI_per_Capita',
    data.columns[7]: 'GNI_Rank_Minus_HDI_Rank',
    data.columns[8]: 'HDI_Rank'
}

data = data.rename(columns=nuevos_nombres)


# Para verificar el resultado (opcional)
print(data.shape)
data.head(15)
```
